1. Classification of objects (species)

2. Classification into a number of groups

3. how to determine the nbr of clusters ? number of outputs , number of closed cross areas

4. Means : Averages : center of each cluster

# Algorithm

1. Random K

2. Classification of each object into a cluster , depending on the closed center to this point

3. Re_clzssifcation (optimization --> calculations of the distance to the centers)

4. Finzl measurment of K clusters

```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
model=KNeighborsClassifier().fit(X_train,y_train)
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("The Accuracy of KNN is: ",accuracy_score(y_test,y_pred)*100," %")
print("The classification Report of KNN is: \n",classification_report(y_test,y_pred))
print("The confusion matrix of KNN is: \n",confusion_matrix(y_test,y_pred))
```

# Supervised learning:

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
model=SVC().fit(X_train,y_train)
y_pred=model.predict(X_test)
#from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("The Accuracy of SVM is: ",accuracy_score(y_test,y_pred)*100," %")
print("The classification Report of SVM is: \n",classification_report(y_test,y_pred))
print("The confusion matrix of SVM is: \n",confusion_matrix(y_test,y_pred))
```

```python
model=RandomForestClassifier().fit(X_train,y_train)
y_pred=model.predict(X_test)
#from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("The Accuracy of Random Forest Classifier is: ",accuracy_score(y_test,y_pred)*100," %")
print("The classification Report of Random Forest Classifier is: \n",classification_report(y_test,y_pred))
print("The confusion matrix of Random Forest Classifier is: \n",confusion_matrix(y_test,y_pred))
```
